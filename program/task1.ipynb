```python
data = pd.read_csv('task1_1.csv')
```

```python
data.销售日期 = pd.to_datetime(data.销售日期,format='%Y%m%d')
data.head()
```

```python
import numpy as np
np.max(data.销售日期) - np.min(data.销售日期)
```

```python
print('共有{}周'.format(119/7))
```

```python
np.min(data.销售日期)
```

```python
np.max(data.销售日期)
```

```python
data['周'] = 0
```

```python
days = {1:31,2:28,3:31,4:30}
ddd = [0,31,59,90]
for i in [1,2,3,4]:
    for j in range(1,days[i]+1):
        boo = (data.销售日期.dt.month == i) & (data.销售日期.dt.day == j)
        if (j+ddd[i-1])%7 == 0:
            w = (j+ddd[i-1])//7
        else:
            w = (j+ddd[i-1])//7 +1
        data.loc[boo,'周'] = w#(j+ddd[i-1])//7 + 1#int((j+ddd[i-1])//7+1)
data.head()
```

```python
data1 = data[(data.商品类型=='生鲜') | (data.商品类型=='一般商品')]
data2 = data1[['商品类型','周','销售金额']].groupby(by=['商品类型','周']).sum()
data3 = data2.unstack()
data3.index.name = ''
data3.columns.name = ''
data3.columns = range(1,19)
data3.to_csv('task1_4.csv')
data3
```

## 任务1.5

```python
data = pd.read_csv('task1_1.csv')
class_peplo = list(data['顾客编号'].unique())
month=[201501,201502,201503,201504]

guke = []
for mon in month:
    gu = []
    day = []
    data0 = data.loc[data.销售月份==mon]
    for cl in class_peplo:
        data1 = data0.loc[data0.顾客编号==cl].销售金额.sum()
        data1 = round(data1,2)
        data2 = len(set(data0.loc[data0.顾客编号==cl].销售日期))
        gu.append(data1)
        day.append(data2)
    guke.append(gu)
    guke.append(day)
data3 = pd.DataFrame(guke)

column = ['1月份消费总额','1月份消费天数','2月份消费总额','2月份消费天数','3月份消费总额','3月份消费天数','4月份消费总额','4月份消费天数',]
data3 = data3.T
data3.columns = column
data3.head()
```

```python
data3.index.name = '顾客编号'
data3.to_csv('task1_5.csv',encoding='gbk')
```
