```python
import pandas as pd
data = pd.read_csv('task1_1.csv')
```

```python
a = data[data.是否促销 == '是'].shape[0]
print('促销商品的购买次数为{}'.format(a))

b = data[data.是否促销 == '否'].shape[0]
print('促销非商品的购买次数为{}'.format(b))
```

```python
a0 = data[data.是否促销 == '是'].销售金额.sum()
print('促销商品的销售总额为{}'.format(a0))

b0 = data[data.是否促销 == '否'].销售金额.sum()
print('非促销商品的销售总额为{}'.format(b0))
```

```python
print('促销商品中，平均每次购买的金额为：{:.2f}'.format(a0/a))
print('非促销商品中，平均每次购买的金额为：{:.2f}'.format(b0/b))
```

```python
data1 = data[['大类名称','是否促销','销售金额']].groupby(by=['大类名称','是否促销']).sum()
```

```python
data1
```

```python
big_kinds = data[data.是否促销=='是']['大类名称'].unique()
print('有进行促销的大类商品有：',big_kinds)
```

```python
sheet = {}
for kind in big_kinds:
    data0 = data[data.大类名称==kind]
    a = data0[data0.是否促销=='是'].销售金额.sum()
    b = data0[data0.是否促销=='否'].销售金额.sum()
    a0 = data0[data0.是否促销=='是'].shape[0]
    b0 = data0[data0.是否促销=='否'].shape[0]
    sheet[kind] = [a,b,round(a/a0,2),round(b/b0,2)]
data2 = pd.DataFrame(sheet)
index = ['促销商品销售总额','非促销商品销售总额','促销商品平均每次购买金额','非促销商品平均每次购买金额']
data2.index = index
data2
```

```python
data3 = data2.T
data3['促销活动后均次消费金额增长率'] = (data3.促销商品平均每次购买金额-data3.非促销商品平均每次购买金额)/data3.非促销商品平均每次购买金额
data3.促销活动后均次消费金额增长率 = data3.促销活动后均次消费金额增长率.apply(lambda x:round(x,4)*100)
data3 = data3.sort_values(by='促销活动后均次消费金额增长率',ascending=False)
data3
```

```python
plt.figure(figsize=(12,8))
plt.bar(range(10),data3.促销活动后均次消费金额增长率,color='r')
plt.xlabel('大类名称',fontsize=20)
plt.ylabel('促销活动后均次消费增长率',fontsize=20)
plt.title('各类商品促销活动后均次消费增长率直方图',fontsize=20)
L = list(data3.促销活动后均次消费金额增长率)
plt.xticks(range(10),data3.index,fontsize=20)
for i in range(10):
    plt.text(i,L[i],round(L[i],2),fontsize=20,va='bottom', ha='center')
plt.savefig('各类商品促销活动后均次消费增长率直方图.png')
```

* 由以上直方图可知，除了洗化类商品和家居类商品均次消费金额对促销活动负敏感，其他大类商品，均对促销活动正敏感，其中粮油类商品为最，促销活动可使得其均次消费提升100%以上。经营策略建议：除了洗化类商品和家居类商品，其他商品皆可进行促销活动，建议优先对粮油类商品、日配类商品、文体类商品进行促销，会取得最好的促销效果

```python
plt.figure(figsize=(12,8))
plt.bar(range(0,20,2),data2.iloc[2,:],color='r',width=0.9)
plt.bar(range(1,21,2),data2.iloc[3,:],color='g',width=0.9)
L = [0.5,2.5,4.5,6.5,8.5,10.5,12.5,14.5,16.5,18.5,20.5]
plt.xticks(L,big_kinds,fontsize=20)
plt.xlabel('大类商品',fontsize=20)
plt.ylabel('平均每次购买金额',fontsize=20)
plt.title('各大类商品促销商品与非促销商品平均每次消费金额对比',fontsize=20)
plt.legend(['促销类商品','非促销类商品'],fontsize=20)
for x in range(0,20,2):
    plt.text(x,data2.iloc[2,x//2],round(data2.iloc[2,x//2],2), va='bottom', ha='center',fontsize=15)
for y in range(1,21,2):
    plt.text(y,data2.iloc[3,y//2],round(data2.iloc[3,y//2],2), va='bottom', ha='center',fontsize=15)
plt.savefig('各大类商品促销商品与非促销商品平均每次消费金额对比.png')
```

* 由以上直方图可知，日配类商品、粮油类商品、酒饮类商品、文体类商品、家电类商品的平均每次消费金额对促销活动较为敏感，当有促销活动时，这几大类的商品均次消费金额比非促销时有较大提升。因此给出的经营策略是优先对这几类商品进行促销，可提升顾客单次消费的金额。而洗化类商品对促销活动不敏感，故无需进行不必要的促销活动。此外，家居类商品对促销活动的敏感度为负，即促销活动使得家居类商品均次消费金额下降，此处建议家居类商品不宜进行促销活动
